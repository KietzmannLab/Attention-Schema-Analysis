```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# load results
returns_per_noise = pd.read_csv('results/returns_per_noise.csv', decimal=',', delimiter=';', index_col=0)
ci_errors = pd.read_csv('results/confidence_intervalls_errors.csv', delimiter=';', index_col=0)
decoding_results = pd.read_csv('results/decoding_results.csv', delimiter=';', decimal=',', index_col=0)
noises = [0, 0.25, 0.5, 0.75, 1.0]
x = np.array(noises)

```

```python
yerr_normal_att = ci_errors['yerr_normal_att'].values
yerr_random_att = ci_errors['yerr_random_att'].values
yerr_normal_catch = ci_errors['yerr_normal_catch'].values
yerr_random_catch = ci_errors['yerr_random_catch'].values

means_att_ci = returns_per_noise['means_att'].values
means_catch_ci = returns_per_noise['means_catch'].values
means_att_random_schema_ci = returns_per_noise['means_att_random_schema'].values
means_catch_random_schema_ci = returns_per_noise['means_catch_random_schema'].values
```

```python

plt.figure()
plt.bar(x-0.025, means_att_ci, width=0.05, label="Learned Additional Resource", yerr=yerr_normal_att,  capsize=3,bottom=0)
plt.bar(x+0.025, means_att_random_schema_ci, width=0.05, color="r", label="Random Additional Resource", yerr=yerr_random_att,  capsize=3, bottom=0)
plt.xticks(x, fontsize = 13)
plt.yticks(fontsize = 13)
plt.xlabel("Noise Probability", fontsize = 13)
plt.ylabel("Ball Tracking Reward (TR)", fontsize = 13)
plt.ylim(0,5.1)
plt.axhline(y=4, color='gray', linestyle='--')
plt.text(0.15, 4.02, 'max', color='gray', fontsize=12, ha='center', va='bottom')

plt.legend(fontsize = 13)
```

```python
bottom = -0.5
means_catch_ci = np.array(means_catch_ci) - bottom
means_catch_random_schema_ci = np.array(means_catch_random_schema_ci) - bottom
plt.figure()
plt.bar(x-0.025, means_catch_ci, width=0.05, label="Learned Additional Resource", yerr=yerr_normal_catch,  capsize=3, bottom=bottom)
plt.bar(x+0.025, means_catch_random_schema_ci, width=0.05, color="r", label="Random Additional Resource", yerr=yerr_random_catch,  capsize=3,bottom=bottom)
plt.xticks(x, fontsize = 13)
plt.yticks(fontsize = 13)
plt.xlabel("Noise Probability", fontsize = 13)
plt.ylabel("Catching Ball Reward (CR)", fontsize = 13)
plt.ylim(-0.5,2.7)
plt.axhline(y=2, color='gray', linestyle='--')
plt.text(0.15, 2.03, 'max', color='gray', fontsize=12, ha='center', va='bottom')


plt.legend(fontsize = 13)
```

```python

cv_aw_col_performance_vis_input = decoding_results['cv_aw_col_performance_vis_input'].values
cv_aw_col_performance_vis_input_schema_pos = decoding_results['cv_aw_col_performance_vis_input_schema_pos'].values
cv_aw_col_performance_from_schema = decoding_results['cv_aw_col_performance_from_schema'].values
cv_aw_col_performance_from_random_schema = decoding_results['cv_aw_col_performance_from_random_schema'].values


```

```python
plt.figure()
plt.bar(x-0.025, cv_aw_col_performance_vis_input_schema_pos, width=0.05, label="Visual Input and Attention Schema Input")
plt.bar(x+0.025, cv_aw_col_performance_vis_input, width=0.05, color="r", label="Only Visual Input")

plt.xticks(x, fontsize = 13)
plt.xlabel("Noise Probability", fontsize = 13)
plt.ylabel(" Attention Window Column Classification Accuracy", fontsize = 13)
plt.yticks(fontsize = 13)
plt.ylim(0,1.3)
plt.legend(fontsize = 13)
```
